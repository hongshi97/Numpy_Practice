### 배열 생성 및 초기화

* `zeros()`: 모든 요소를 0으로 초기화

```
np.zeros(10)
```

* `ones()`: 모든 요소를 1로 초기화

```
print(np.ones(10))
print('='*30)
print(np.ones((3,3)))
```

* `full()`: 모든 요소를 지정한 값으로 초기화

```
np.full((3,3), fill_value = 1.23)
```

* `eye()`: 단위행렬(identity matrix) 생성
  + 주대각선의 원소가 모두 1이고 나머지 원소는 모두 0인 정사각 행렬

```
np.eye(4) # 단위 행렬은 정사각 행렬이라 shape 모양을 정해주는 게 아니라, n 값만 넣으면 n x n 정사각 행렬 만들 수 있음
```

* `tri()`: 삼각행렬 생성

```
np.tri(4) # np.tri()도 np.eye() 처럼 n 값만 넣으면 됨
```

* `empty()`: 초기화되지 않은 배열 생성
  + 초기화가 없어서 배열 생성비용 저렴하고 빠름
  + 초기화되지 않아서 기존 메모리 위치에 존재하는 값이 있음

```
print(np.empty((10)))  # 주의! 초기화를 하지 않았는데 출력 결과를 보면 값이 1.로 채워져있음.  기존 메모리 위치에 존재하는 값이 있어서 이렇게 불러와진 거
print('='*30)
print(np.empty((5,5))) # 5 x 5 array도 이미 특정한 값으로 채워져있네. 즉, np.empty()로 임의의 배열을 생성할 때에는 안에 element 값이 어떠한 값으로 채워져있을지 모르니까 주의!!! 
```

* `_like()`: 지정된 배열과 shape가 같은 행렬 생성
  + `np.zeros_like()`
  + `np.ones_like()`
  + `np.full_like()`
  + `np.empty_like()`

```
print(a1)

print(np.zeros_like(a1))
```

```
np.ones_like(a1)
```

```
np.full_like(a1)  # np.full()과 np.full_like()는 fill_value 값을 입력해줘야 함!
```

```
np.full_like(a1, fill_value = 5)
```

### 인덱싱(Indexing)

```
print(a1)
print(a1[0])
print(a1[-1])
print(a1[-3])
```

```
print(a2)
print('='*30)
print(a2[0])
print(a2[0,0]) # Numpy Array와 파이썬 List 형태의 차이! 이중 List의 경우 Element 개별 값에 접근하기 위해서는 list[n,m]은 에러가 뜨고 list[n][m]과 같이 인덱싱해야 하는데 Numpy Array는 array[n,m]으로 인덱싱 가능
print(a2[-1,1])
print(a2[-2,-2])
```

```
print(a3)
print('='*30)
print(a3[0,0,0])
print(a3[1,1,1])
print(a3[2,2,2])
print(a3[2,-1,-1])
```

### 슬라이싱(Slicing)

* 슬라이싱 구문: `a[start:stop:step]`
* 기본값: start=0, stop=ndim, step=1

```
print(a1)
print('='*30)
print(a1[0:2])
print(a1[0:])
print(a1[:1]) # start 값이 없으면 자동으로 index 0 부터 시작함
print(a1[::2]) # start랑 stop 지정 안하고 step = 2
print(a1[::-1]) # start랑 stop 지정 안하고 step = -1 하면 원래 array의 역순을 반환
```

```
print(a2)
print('='*30)
print(a2[1])
print('-'*30)
print(a2[1, :])
print('-'*30)
print(a2[:2, :2])
print('-'*30)
print(a2[1:, ::-1])
print('-'*30)
print(a2[::-1, ::-1])
```

### 불리언 인덱싱(Boolean Indexing)

* 배열 각 요소의 선택 여부를 불리언(True or False)로 지정
* True 값인 인덱스의 값만 조회

```
print(a1)
bi = [False, True, True, False, True]
print(a1[bi])
bi2 = [True, False, True, True, False]
print(a1[bi2])
```

```
print(a2)
bi = np.random.randint(0, 2, (3,3), dtype = bool)
print(bi)
print(a2[bi]) # 주의! 2차원 array에 불리언 인덱싱을 적용한 결과는 1차원 array로 나오네
```

### 팬시 인덱싱(Fancy Indedxing)

```
print(a1)
print([a1[0],a1[2]])

idx = [0,2]
print(a1[idx])

idx2 = np.array([[0,1],   # 주의! index 형태에 따라 팬시 인덱싱 결과값 형태가 달라짐,
                 [2,0]])
print(a1[idx2])           # 주의! idx2가 만약에 array가 아니라 그냥 List 자료형이면 a1[idx2] 에러 발생함 
```

```
print(a2)
print('='*30)
row = np.array([0,2])
col = np.array([1,2])
print(a2[row,col])   # row = 0, col = 1인 거, row = 2, col = 2인 거 하나씩 반환
print('-'*30)
print(a2[row, :])
print('-'*30)
print(a2[:,col])
print('-'*30)
print(a2[row, 1])
print('-'*30)
print(a2[1,col])
print('-'*30)
print(a2[row, 1:])
print('-'*30)
print(a2[1:, col])
```

## 배열 값 삽입/수정/삭제/복사

### 배열 값 삽입

* `insert()`: 배열의 특정 위치에 값 삽입
* axis를 지정하지 않으면 1차원 배열로 변환
* 추가할 방향을 axis로 지정
* 원본 배열 변경없이 새로운 배열 반환

```
print(a1)
b1 = np.insert(a1, 0, 10)  # a1이라는 array의 index 0에 10이라는 값 추가 

print(a1)
print(b1) # insert()는 원본 배열 변경을 하지 않고 새로운 배열을 반환함

c1 = np.insert(a1, 2, 10)
print(c1)
```

```
print(a2)
print('-'*30)
b2 = np.insert(a2, 1, 10, axis = 0)
print(b2)
print('-'*30)
c2 = np.insert(a2, 1, 10, axis = 1)
print(c2)
```

### 배열 값 수정

* 배열의 인덱싱으로 접근하여 값 수정

```
print(a1)
a1[0] = 1
a1[1] = 2
a1[2] = 3
print(a1)

a1[:2] = 9
print(a1)

i = np.array([1,3,4])  # 팬시 인덱싱 활용
a1[i] = 0
print(a1)

a1[i] += 4
print(a1)
```

```
print(a2)
print('-'*30)

a2[0,0] = 1
a2[1,1] = 2
a2[2,2] = 3
print(a2)
print('-'*30)

a2[0] = 1
print(a2)
print('-'*30)

a2[1:,2] = 9  # 슬라이싱 활용
print(a2)
print('-'*30)

row = np.array([0,1])  # 팬시 인덱싱 활용
col = np.array([1,2])
a2[row,col] = 0
print(a2)
```

- np.where() 사용
1. **np.where(조건, True일 때 값, False일 때 값)**
2. 벡터 연산을 하므로 for loop를 돌지 않아서 속도가 매우 빠름
3. 원래 Array는 변경되지 않음

```
temp_array = np.arange(1,28).reshape(3,3,3)
print(temp_array)
print('-'*30)

temp_array2 = np.where(temp_array**2 <= 16, 1000, temp_array)  # 제곱값이 16이하인 수는 1000으로, 나머지는 그대로 변환.
print(temp_array2)
```

### 배열 값 삭제

* `delete()`: 배열의 특정 위치에 값 삭제
* axis를 지정하지 않으면 1차원 배열로 변환
* 삭제할 방향을 axis로 지정
* 원본 배열 변경없이 새로운 배열 반환

```
print(a1)
b1 = np.delete(a1, 1)
print(b1)
print(a1) # 원본 배열은 변경 X
```

```
print(a2)
print('-'*30)

b2 = np.delete(a2, 1, axis = 0)
print(b2)
print('-'*30)
print(a2)

print('-'*30)
c2 = np.delete(a2, 1, axis = 1)
print(c2)
```

### 배열 복사

* 리스트 자료형과 달리 배열의 슬라이스는 복사본이 아님

```
print(a2)
print('-'*30)
print(a2[:2, :2])
print('-'*30)
a2_sub = a2[:2,:2]
print(a2_sub)
print('-'*30)
a2_sub[:,1] = 0
print(a2_sub)
print('-'*30)
print(a2)    # Numpy Array에서 슬라이싱을 해서 새로운 변수를 수정하면 원본 배열에도 영향이 있구나
```


* `copy()`: 배열이나 하위 배열 내의 값을 명시적으로 복사

```
print(a2)
print('-'*30)

a2_sub_copy = a2[:2,:2].copy()
print(a2_sub_copy)
print('-'*30)

a2_sub_copy[:,1] = 1
print(a2_sub_copy)
print('-'*30)

print(a2)  # copy()를 이용했더니, 원본은 변하지 않음
```

## 배열 연산

* NumPy의 배열 연산은 벡터화(vectorized) 연산을 사용
* 일반적으로 NumPy의 범용 함수(universal functions)를 통해 구현
* 배열 요소에 대한 반복적인 계산을 효율적으로 수행

### 브로드캐스팅(Broadcasting)

### 집계 함수(Aggregate Functions)

#### sum(): 합 계산

```
a2 = np.random.randint(1,10, size = (3,3))
print(a2)
print('-'*30)

# axis 지정 X: axis를 지정하지 않고 sum()을 하면 전체 element의 합을 계산
print(a2.sum(), np.sum(a2))
print('-'*30)

# axis 지정: axis에 따라 element 합 계산
print(a2.sum(axis = 0), np.sum(a2,axis = 0))
print('-'*30)

print(a2.sum(axis = 1), np.sum(a2, axis = 1))
```

#### cumsum(): 누적합 계산

```
print(a2)
print('-'*30)

# axis 지정 X: array 안 모든 element의 합 계산
print(a2.cumsum(), np.cumsum(a2))
print('-'*30)

# axis 지정: axis에 따라 element 합 계산
print(a2.cumsum(axis = 0))
print('-'*30)
print(np.cumsum(a2,axis = 0))
print('-'*30)

print(a2.cumsum(axis = 1))
print('-'*30)
print(np.cumsum(a2, axis = 1))
```

#### diff(): 차분 계산

```
print(a2)
print('-'*30)

# diff()는 a2.diff() 불가

# axis 지정 X: default는 axis = 1
print(np.diff(a2))
print('-'*30)

# axis 지정: axis에 따라 element 합 계산
print(np.diff(a2,axis = 0))
print('-'*30)

print(np.diff(a2, axis = 1))
```

#### prod(): 곱 계산

```
print(a2)
print('-'*30)

print(np.prod(a2))
print('-'*30)

print(np.prod(a2, axis = 0))
print('-'*30)

print(np.prod(a2, axis = 1))
```

#### cumprod(): 누적곱 계산

```
print(a2)
print('-'*30)

print(np.cumprod(a2))
print('-'*30)

print(np.cumprod(a2, axis = 0))
print('-'*30)

print(np.cumprod(a2, axis = 1))
```

#### 불리언 연산자(Boolean Operators)


image.png

```
a2 = np.arange(1,10).reshape(3,3)
print(a2)
print('-'*30)

print((a2 > 5) & (a2 < 8))
print('-'*30)

print(a2[(a2 > 5) & (a2 < 8)])
print('-'*30)

print((a2 > 5) | (a2 < 8))
print('-'*30)

print(a2[(a2 > 5) | (a2 < 8)])
print('-'*30)

print((a2 > 5) ^ (a2 < 8))  # XOR 연산: 입력값이 같이 않으면 True(1)   Ex) 0 XOR 0 = 0,  0 XOR 1 - 1,  1 XOR 0 = 1,  1 XOR 1 = 0
print('-'*30)

print(a2[(a2 > 5) ^ (a2 < 8)])
print('-'*30)

print(~(a2 > 5))  
print('-'*30)

print(a2[~(a2>5)]) # ~(a2>5) : a2>5가 아닌 것들 즉, a2 <= 5인 것들. (~는 not의 역할)
```

### 배열 정렬

```
a1 = np.random.randint(1,10, size = 10)
print(a1)
print('-'*30)

print(np.sort(a1))
print(a1) # 원본 배열은 변하지 않고 그대로!
print('-'*30)

print(np.argsort(a1))  # argsort()는 index 값. a1 array에서 어느 index에 있는 element가 이 자리에 와야 하는지
print(a1) # 원본 배열은 변하지 않고 그대로!
print('-'*30)

print(a1.sort())  # np.sort(array)가 아니라 array.sort()를 해주면 원본 배열이 변형됨
print(a1)
```

```
a2 = np.random.randint(1,10, size = (3,3))
print(a2)
print('-'*30)

print(np.sort(a2, axis = 0))
print('-'*30)

print(np.sort(a2, axis = 1))
```
